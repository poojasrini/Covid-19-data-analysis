# ***COVID-19 DATA ANALYSIS - INDIA***

The outbreak of COVID-19 is all across the World and is been spreading day by day. Thus the correct analysis to predict the cases in future along with finding the correlation by plotting graphs is very essential for clear understanding.
<br/>

The objective of this notebook is to understand the outbreak of COVID-19 in India, Comparison of different states in India, and finding the worst affected state in this pandemic. A Machine Learning and Time Series Forecasting model is also built to understand what these numbers are going to be like in near future.

The dataset was obtained from kaggle : https://www.kaggle.com/sudalairajkumar/covid19-in-india?select=covid_19_india.csv <br/>
This in turn was obtained form the Indian Ministry of Health & Family Welfare (https://www.mohfw.gov.in/) which is available to general public and has data starting from 30.01.2020 to 16.11.2020 for every state in india.

```
import pandas as pd
import matplotlib.pyplot as plt
india=pd.read_csv('/content/drive/MyDrive/python/covid_19_india.csv')
india.tail()
```

```
india['State/UnionTerritory'].value_counts()
```

### Separating the top 11 states into their respective datasets

```
kerala_data=india[india['State/UnionTerritory']=='Kerala']
tamilnadu_data=india[india['State/UnionTerritory']=='Tamil Nadu']
maharashtra_data=india[india['State/UnionTerritory']=='Maharashtra']
rajasthan_data=india[india['State/UnionTerritory']=='Rajasthan']
delhi_data=india[india['State/UnionTerritory']=='Delhi']
uttarpradesh_data=india[india['State/UnionTerritory']=='Uttar Pradesh']
haryana_data=india[india['State/UnionTerritory']=='Haryana']
karnataka_data=india[india['State/UnionTerritory']=='Karnataka']
madhyapradesh_data=india[india['State/UnionTerritory']=='Madhya Pradesh']
ladakh_data=india[india['State/UnionTerritory']=='Ladakh']
jammuandkashmir_data=india[india['State/UnionTerritory']=='Jammu and Kashmir']

```

## **NUMBER OF ACTIVE CASES**
### Active cases => Number of confirmed cases-Number of recovered cases-Number of deaths

```
plt.plot(kerala_data['Date'],kerala_data["Confirmed"]-kerala_data["Cured"]-kerala_data["Deaths"], label='Kerala')
plt.plot(tamilnadu_data['Date'],tamilnadu_data["Confirmed"]-tamilnadu_data["Cured"]-tamilnadu_data["Deaths"], label='Tamil Nadu')
plt.plot(ladakh_data['Date'],ladakh_data["Confirmed"]-ladakh_data["Cured"]-ladakh_data["Deaths"], label='Ladakh')
plt.plot(maharashtra_data['Date'],maharashtra_data["Confirmed"]-maharashtra_data["Cured"]-maharashtra_data["Deaths"], label='Maharastra')
plt.plot(karnataka_data['Date'],karnataka_data["Confirmed"]-karnataka_data["Cured"]-karnataka_data["Deaths"], label='Karnataka')
plt.plot(haryana_data['Date'],haryana_data["Confirmed"]-haryana_data["Cured"]-haryana_data["Deaths"], label='Harayana')
plt.plot(jammuandkashmir_data['Date'],jammuandkashmir_data["Confirmed"]-jammuandkashmir_data["Cured"]-jammuandkashmir_data["Deaths"], label='Jammu and Kashmir')
plt.plot(uttarpradesh_data['Date'],uttarpradesh_data["Confirmed"]-uttarpradesh_data["Cured"]-uttarpradesh_data["Deaths"], label='Uttar Pradesh')
plt.plot(madhyapradesh_data['Date'],madhyapradesh_data["Confirmed"]-madhyapradesh_data["Cured"]-madhyapradesh_data["Deaths"], label='Madhya Pradesh')
plt.plot(delhi_data['Date'],delhi_data["Confirmed"]-delhi_data["Cured"]-delhi_data["Deaths"], label='Delhi')
plt.title('DATE Vs NUMBER OF ACTIVE CASES')
plt.xlabel('DATES')
plt.ylabel('NUMBER OF ACTIVE CASES')
plt.legend()
plt.rcParams["figure.figsize"] = (15,9)# resizing plot
plt.show()
```

It is seen that **MAHARASHTRA** has the maximum number of active cases. However as a whole the the cases have been decreasing for all the states over time as seen in the graph.

# F**BPROPHET PREDICTION**
## Top 4 affected states prediction

### **MAHARASHTRA**

```
import fbprophet
mah_new = maharashtra_data[['Date','Infection Rate']]
mah_new = mah_new.rename(columns={'Date':'ds','Infection Rate':'y'})
op_mah = fbprophet.Prophet(changepoint_prior_scale=1)
op_mah.fit(mah_new)
forecast_mah = op_mah.make_future_dataframe(periods=12,freq='w')
forecast_mah = op_mah.predict(forecast_mah)
op_mah.plot(forecast_mah,xlabel = 'year-month',ylabel = 'No of cases')
plt.title('Maharashtra forecasting')

```

### **DELHI**

```

del_new = delhi_data[['Date','Infection Rate']]
del_new = del_new.rename(columns={'Date':'ds','Infection Rate':'y'})
op_del = fbprophet.Prophet(changepoint_prior_scale=1)
op_del.fit(del_new)
forecast_del = op_del.make_future_dataframe(periods=12,freq='W')
forecast_del = op_del.predict(forecast_del)
op_del.plot(forecast_del,xlabel = 'year-month',ylabel = 'No of cases')
plt.title('Delhi forecasting')
```

### **TAMIL NADU**

```
tn_new = tamilnadu_data[['Date','Infection Rate']]
tn_new = tn_new.rename(columns={'Date':'ds','Infection Rate':'y'})
op_tn = fbprophet.Prophet(changepoint_prior_scale=1)
op_tn.fit(tn_new)
forecast_tn = op_tn.make_future_dataframe(periods=12,freq='W')
forecast_tn = op_tn.predict(forecast_tn)
op_tn.plot(forecast_tn,xlabel = 'year-month',ylabel = 'No of cases')
plt.title('Tamil Nadu forecasting')

```

### **KERALA**

```

ker_new = kerala_data[['Date','Infection Rate']]
ker_new = ker_new.rename(columns={'Date':'ds','Infection Rate':'y'})
op_ker = fbprophet.Prophet(changepoint_prior_scale=1)
op_ker.fit(ker_new)
forecast_ker = op_ker.make_future_dataframe(periods=12,freq='W')
forecast_ker = op_ker.predict(forecast_ker)
op_ker.plot(forecast_ker,xlabel = 'year-month',ylabel = 'No of cases')
plt.title('Kerala forecasting')
```

From the above graphs it is seen that the number of cases is been going up and down with time but however as a whole its been reducing.
